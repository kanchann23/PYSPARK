```python
from pyspark.sql.functions import col, sum

# Calculate the count of missing values for each column in the DataFrame 'df'
missing_counts = data.agg(*[
    sum(col(c).isNull().cast('int')).alias(c + '_missing') for c in data.columns
])
#cast to change boolean in to integer
#sum will work only integer bcz
# Calculate the total number of rows in the DataFrame
total_rows = data.count()

# Filter the columns with null values
columns_with_null = [c for c in data.columns if missing_counts.select(col(c + '_missing')).collect()[0][0] > 0]

# Print the columns with null values
print("Columns with null values:", columns_with_null)
```

```python
#alias(c + '_missing'):
#The alias() method is used to provide a new name for the column resulting from the aggregation.
#In this case, we add '_missing' to the original column name to indicate that this new column contains the count of missing values for that column.
```

```python
from pyspark.sql.functions import mean

# Calculate the mean of each numeric column
numeric_cols = [c for c, t in data.dtypes if t in ('double', 'int')]
means = data.select(*[mean(col(c)).alias(c) for c in numeric_cols]).first().asDict()

# Impute missing values in numeric columns with their respective means
for col_name in numeric_cols:
    data = data.na.fill(means[col_name], [col_name])

# Impute missing values in non-numeric columns with 'None'
non_numeric_cols = [c for c, t in data.dtypes if t not in ('double', 'int')]
for col_name in non_numeric_cols:
    data = data.na.fill('None', [col_name])

# Show the DataFrame after imputation
data.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert PySpark DataFrame to Pandas DataFrame for plotting
pandas_df = data.toPandas()

# Plot the distribution between gender and dependents
plt.figure(figsize=(8, 6))
pd.crosstab(pandas_df['Gender'], pandas_df['Dependents']).plot(kind='bar', stacked=True)
plt.title('Distribution between Gender and Dependents')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Dependents', loc='upper right')
plt.show()
```
