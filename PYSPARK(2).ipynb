# PYSPARK HANDLING MISSSING DATA

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('pysparkk').getOrCreate()
```

```python
data = spark.read.csv('/home/user/Downloads/pyspark/data.csv',header=True,inferSchema=True)
```

```python
data.show()
```

```python
#drop
data.drop('work').show()
```

```python
# dropping rows having null value
data.na.drop().show()
```

```python
#if any record all row value are null we use na.drop(how="all")
data.na.drop(how="all").show()
```

```python
# any if there is any null
data.na.drop(how="any").show()
```

```python
data.na.drop(how="any",thresh=2).show()
# thresold for minimum value
```

```python
#if I want to drop values only from specific column we use subset
data.na.drop(how="any",subset=['work']).show()
```

```python
# filling missing values 
# value and subset
data.na.fill('missed').show()
```

# filter operation

```python
#age less than70
```

```python
data.filter('age<70').show()
```

```python
# two different condition
```

```python
data.filter((data["age"]<70)&(data['age']>20)).show()
```
